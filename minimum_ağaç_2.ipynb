```python
import networkx as nx
import matplotlib.pyplot as plt

vertices = ['A', 'B', 'C', 'D', 'E', 'F']
edges = [
    ('A', 'B', 4), ('A', 'C', 2), ('B', 'C', 5), ('B', 'D', 10),
    ('C', 'D', 3), ('C', 'E', 8), ('D', 'E', 4), ('D', 'F', 6), ('E', 'F', 1)
]

# Graf oluşturma
G = nx.Graph()
G.add_weighted_edges_from(edges)

#MST
mst = nx.minimum_spanning_tree(G)

print("Minimum Spanning Tree:")
mst_weight = 0
for u, v, data in mst.edges(data=True):
    print(f"{u} - {v} : {data['weight']}")
    mst_weight += data['weight']

print(f"Toplam Ağırlık: {mst_weight}")

# Graf
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Yerleşim algoritması
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=800, node_color='blue', font_size=10, font_color='black', font_family='Arial')


edges_mst = mst.edges()
nx.draw_networkx_edges(G, pos, edgelist=edges_mst, edge_color='red', width=2)

# Grafı gösterme
plt.title("Graf ve Minimum Spanning Tree (Kruskal Algoritması)")
plt.show()
```
